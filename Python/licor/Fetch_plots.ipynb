```python
import xarray as xr
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
file_list = glob.glob('/Users/bhupendra/projects/crocus/data/flux_data/data/netcdf/resnc/*/*.nc')
file_list.sort()
ds = xr.open_mfdataset(file_list)
```

# Max Fetch 

```python
hst = ds.x_peak.plot.hist(bins=100)
plt.xlabel('Max. Fetch [m]')
```

The Max Fetch is coming from close to 200 m, which is the Parking tower distance. 

```python
hst = ds['x_90%'].plot.hist(bins=500, xlim=(0, 1000))
plt.xlabel('90% Fetch [m]')
```

```python

ds.plot.scatter(x='x_peak', y='wind_dir', alpha=0.1)
```

The northern quadrant is very clean when it comes to Fetch. This is parking tower side where there is also a very large interstate junction but unfortunately that is out of fetch area. 

```python
ds.plot.scatter(x='x_peak', y='wind_speed', alpha=0.1)
```

It is very homogeneous in the northern side of the site for up to 500 meters, which may be the reason for such consistent fetch and higher wind speed.
